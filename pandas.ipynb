# Video 5: Modifying Data within DataFrames

```python
#suppose you are given this small data set
people = {
    "first": ["Ali", "Ehap", "Youssef", "Haneen"],
    "second": ["Ahmed", "Saad", "Nagy", "Amin"],
    "last": ["aa@email.com", "bb@email.com", "cc@email.com", "dd@email.com"]
}
df = pd.DataFrame(people)
#see how the given column names are not meaningful
df
```

```python
#let's say you want to modify the names of all the columns
df.columns = ['first_name', 'last_name', 'email'] #if the length of list is less than the number of columns, an
#error shows up
df
```

```python
#what if you want to make the names of your columns all uppercase?
df.columns = [col.upper() for col in df.columns]
df
```

```python
#now say you want to replace each _ in the column names with spaces. df.columns series has an str method
df.columns = df.columns.str.replace('_', ' ')
df
```

```python
#so far we have seen ways to update all the columns at once. but what if you want to change a specefic column?
# df.rename(columns={'old_col_name': 'new_col_name', '...': '...', ...})
#this is also a method that requires using inplace attribute

df.rename(columns={'LAST NAME': 'last_name', 'FIRST NAME': 'first_name'}, inplace=True)
df
```

```python
# now that we have looked at how to update columns, we move our attentoin to rows.
#remember how we accessed a row using loc and iloc? we can update an entire row in the same manner
df.loc[2] = ['Ussef', 'Saad', 'yy@email.com']
df
```

```python
# what if you want to update just the email and first name of row 2? df.loc[row_name, col_name]
#df.loc[2, ["first_name", 'email']] = ["Youssef", 'cc@email.com']
df.loc[2, ["first_name", 'EMAIL']] = ["Youssef", 'cc@email.com']
#notice that if the name of the provided col is not in the data frame, pandas will create a new col with that name
df
```

```python
#in case you want to update a single attribute within a row, it is better to use .at[] but loc[] works too
df.at[2, 'last_name'] = 'Nagy'
df
```

```python
#task: rename make the emails in the EMAIL column all uppercase
df['EMAIL'] = df['EMAIL'].str.upper() #df['EMAIL'] returns a series. apply str method on it
df
```

```python
#there are 4 more ways to update multiple rows at once:
# 1- apply   2- map   3- applymap   4- replace
#we will start with apply: it is used to apply a function on our values. apply can be used with series and
#DataFrame objects but its behavoiur will be different in each case
#task: find the length of the first names in each row in one line
df['first_name'].apply(len) #apply the len function on the first_name column
```

```python
#make the emails all in lower case using apply
def to_up(x):
    return x.upper()
df['EMAIL'] =  df['EMAIL'].apply(to_up)
```

```python
df
```

```python
#in case of simple functions like these, we can use lambda function instead
df['EMAIL'] = df['EMAIL'].apply(lambda x: x.lower())
df
```

```python
#so far we have used apply on series objects. but what about DataFrame objects?
# using apply on a DataFrame will apply it to each series in the DataFrame
df.apply(len)
```

```python
#we now move to applymap() this method only applies to DataFrames NOT series objects
# we have seen that using apply on a DataFrame will apply it to each series object. 
#applymap() will apply the passed in function to each element in each series
df.applymap(len)
```

```python
#task: make each element in the DataFrame upper case
df = df.applymap(str.upper)
```

```python
df
```

```python
df['age'] = [10, 20, 25, 30] #make a new column called age
df
```

```python
#now what if we want to make all the strings in the DataFrame lower case? applying str.lower() will cause an
#error because of the integer values
def to_lower(x):
        try: x = x.lower()
        except: pass
        return x

df = df.applymap(to_lower)
df
```

```python
#3- map(). map is used to substitute values within a series with other values. it is only applicable to series
#objects

df['first_name'].map({'haneen':'ahmed'}) 
#notice that the values that were not specified are substituted with NaN value
```

```python
#to avoid the problem with map function we can use replace() instead
df['first_name'] = df['first_name'].replace({'haneen':'ahmed'})
df
```

# Video 9: Cleaning Data

```python
import pandas as pd
import numpy as np

people = {
    'first': ['ali', 'ahmed', 'mohammed', np.nan, 'hamada', 'NA'],
    'last': ['saad', 'adam', 'zayn', np.nan, np.nan, 'Missing'],
    'email': ['aa@email.com', 'bb@email.com', 'cc@email.com', None, np.nan, 'NA'],
    'age': ['33', '55', '56', None, None, 'Missing']
}

df = pd.DataFrame(people)
df
```

```python
#cleaning data may require the any nan values be removed
#removing nan values is achieved via the dropna() method
#we will first run dropna with the default args
df.dropna()
#notice how all the rows that contained nan values were removed
#also notice that the custom nan values that we added were not removed. we will see how to handle them
```

```python
#the default args are dropna(axis='index', how='any') this means remove the rows (axis='index') in which at least one nan value is present (how='any')
#we could alternatively remove the columns that contain nan values
df.dropna(axis='columns', how='any')
#since all the columns contain nan values and how was set to any, all the columns will be removed
```

```python
#suppose we want to remove the columns in which all the values are nan
df.dropna(axis='columns', how='all')
#no col will be removed since no col contains all-nan values
```

```python
#how about we remove the rows in which all values are nan
df.dropna(axis='index', how='all')
#rows 3, 4 are all-nan values. they will be removed
#row 5 is also all-nan but it is our custom nan values
```

```python
#suppose that you want to drop missing values in a specific column.
#drop the rows that are missing emails
df.dropna(subset=['email'])
```

```python
#drop the rows that are missing emails or first names
df.dropna(how='any', subset=['last', 'email'])
```

```python
#drop the rows that are missing both first name and email
df.dropna(how='all', subset=['first', 'email'])
```

```python
#now what about the customized missing values? 
#since we entering the data manually from a dict, we can apply the replace() method on the data fram
temp_df = df.replace(['Missing', 'NA'], np.nan)
#all the custom missing values are replaced with nan now
temp_df
```

```python
temp_df.dropna()
```

```python
#you can also apply the isna() method to the data frame to find where the missing values are
df.isna()
```

```python
temp_df.isna()
```

```python
#suppose instead of deleting na values, you want to replace them with a value of say 0
df.fillna(0)
```

```python
temp_df.fillna(0)
```

```python
#remember that our df data frame has some missing values in the age column. what if we want to calculate the average age?
#nan values are considered float data types. we can simply typecast the age column to float data type
#also remember that we have some customized missing values in this col which needs special treatment

df['age'] = df['age'].replace({'Missing': np.nan})
df['age'] = df['age'].astype(float)
df['age'].mean()
```

```python
import pandas as pd

df = pd.read_csv('ETH_1h.csv')
df_temp = df.copy()
df
```

```python
df.info()
#notice how the Date column is of type str. we want to convert it to a daytime object to carry out some caculations
```

```python
#if we apply the to_datetime on the current Date series we will get an error since this time formatting is not recognized by pandas
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#what we can do is to pass the format of our date to the to_datetime method by the attribute format
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %I-%p')
#%Y-%m-%d %I-%p is the formatting of the date in our dataset
df['Date']
#now the date column is of type datetime
```

```python
#there is another in which we can use apply method and create another col with the dates only.
df_temp['Dates_only'] = df_temp['Date'].apply(lambda x: x[0:10])
df_temp
```

```python
#we can now apply datetime methods to our dataset
df.loc[0, 'Date'].day_name()
```

```python
#find the day name of the entire Date column
df['Day'] = df['Date'].apply(lambda x: x.day_name())
df
```

```python
#there is another way in which we access the dt class on the series
df['Date'].dt.day_name()
```

```python
#find the number of days between the earliest and the latest date
df['Date'].max() - df['Date'].min()
```

```python
df.groupby(by='Day').get_group("Friday")
```

```python
days = ['Friday', 'Monday', 'Sunday']
filt = (df['Date'].dt.year > 2019) & (df['Day'].isin(days))
df[filt]

#we can do this too: filt = (df['Date'] > '2019') & (df['Day'].isin(days))

```

```python
#find the average closing price for 01-2020 and 02-2020

df['YearAndMonth'] = df['Date'].apply(lambda x: str(x.year) + '-' + str(x.month))
df
```

```python
filt = (df['YearAndMonth'] == '2020-2') | (df['YearAndMonth'] == '2020-1')
df[filt]['Close'].mean()
```

```python
#notice how the data in the dataset are sampled hourly. what if we want to resample it on a daily basis?
#the resample() method is only applicable to DatetimeIndex, TimedeltaIndex or PeriodIndex. this means that the index of the data frame must 
#be one of those
df = df.set_index('Date')
df
```

```python
highs = df['High'].resample('D').max() #resample the High col on a daily basis then find the max in each day
```

```python
%matplotlib inline
```

```python
highs.plot()
```

```python
#task: resample the entire data frame on a weekly basis then find the mean of the closing price, the max of the high column, the min of the low col
#and the sum of the volume col

df.resample('W').agg({'Close': 'mean', 'High': 'max', 'Low': 'min', 'Volume': 'sum'})
#this will return the average closing price, the max high, the min low and the volume sum in each week 
```
